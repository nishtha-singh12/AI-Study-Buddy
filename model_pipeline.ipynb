### **MODEL TRAINING**

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
ad.head()
```

```python
le = LabelEncoder()
ad["gender"] = le.fit_transform(ad["gender"])
ad["part_time_job"] = le.fit_transform(ad["part_time_job"])
ad["diet_quality"] = le.fit_transform(ad["diet_quality"])
ad["parental_education_level"] = le.fit_transform(ad["parental_education_level"])
ad["extracurricular_participation"] = le.fit_transform(ad["extracurricular_participation"])
```

```python
ad_model = ad.copy()
```

```python
ad_model
```

```python
# Features and target
X = ad[[
    "gender", "study_hours_per_day", "attendance_percentage",
    "mental_health_rating", "sleep_hours", "diet_quality",
    "part_time_job","parental_education_level", "extracurricular_participation", "social_media_hours"]]
Y = ad["exam_score"]

#Train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

#Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train,Y_train)
y_pred_lr = lr_model.predict(X_test)

#Decision Tree Regressor
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train,Y_train)
y_pred_dt = dt_model.predict(X_test)

#Random Forest Regressor
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train,Y_train)
y_pred_rf = rf_model.predict(X_test)


#Evaluate Linear Regression
mse = mean_squared_error(Y_test, y_pred_lr)
rmse = np.sqrt(mse)
print("Linear Regression")
print("r2", r2_score(Y_test, y_pred_lr))
print("mse:", mse)
print("rmse:", rmse)
print("------")

# Evaluate Decision Tree Regressor
mse = mean_squared_error(Y_test, y_pred_dt)
rmse = np.sqrt(mse)
print("Decision Tree Regressor")
print("r2:", r2_score(Y_test, y_pred_dt))
print("mse:", mse)
print("rmse:", rmse)
print("------")

# Evaluate Random Forest Regressor
mse = mean_squared_error(Y_test, y_pred_rf)
rmse = np.sqrt(mse)
print("Random Forest Regressor")
print("r2:", r2_score(Y_test, y_pred_rf))
print("mse:", mse)
print("rmse:", rmse)
print("------")

```

```python
model_results = {
    "Linear Regression":{
        "r2": 0.8201698053012326,
        "mse": 46.113694260108915,
        "rmse": 6.7907064625198545
    },

    "Decision Tree Regressor":{
        "r2": 0.5431534415267758,
        "mse": 117.14875,
        "rmse": 10.8235276134909
    },

    "Random Forest Regressor":{
        "r2": 0.80383606480444,
        "mse": 50.30214057000001,
        "rmse": 7.092400198099372
    }
}
```
